```python
import ipywidgets
import numpy as np
import matplotlib.pyplot as plt

# Plot Function
def plot_fct(w=1):
  x = np.random.uniform(0, 5, size=100)
  y = 2 * x + w * np.random.normal(size=100)
  plt.scatter(x, y)

#semakin besar nilai w, semakin tidak kolinear grafiknya

plot_fct(10)
```

```python
# Interactive Plot
ipywidgets.interact(plot_fct, w=(0, 5, 0.5))
```

```python
# Randomization
x = np.random.uniform(0, 5, size=100)
noise = np.random.normal(size=100)

def plot_fct(w=1):
  y = 2 * x + w * noise
  plt.scatter(x, y)

ipywidgets.interact(plot_fct, w=(0, 5, 0.1))
```

```python
# MakeMoons
from sklearn.datasets import make_moons

moons = make_moons(n_samples=500, noise=0.05) # semakin besar nilai sampel dan nilai noise, semakin ramai grafiknya
x, y = moons[0], moons[1]

plt.scatter(x[:, 0], x[:, 1], c=y)
```

### Penjelasan
- `make_moons`: Fungsi ini menghasilkan dataset dua dimensi berbentuk bulan sabit yang sering digunakan untuk tugas klasifikasi.
- `n_samples=100`: Menentukan jumlah sampel yang ingin kita hasilkan.
- `noise=0.1`: Menentukan tingkat kebisingan (noise) yang ditambahkan pada data.
- `x` dan `y`: `x` adalah array dua dimensi dari fitur (koordinat), dan `y` adalah array satu dimensi dari label (klasifikasi).

```python
# Interactive MakeMoons
def plot_moons(samples=200, noise=0):
  moons = make_moons(n_samples=samples, noise=noise, random_state=50)
  x, y = moons[0], moons[1]

plt.scatter(x[:, 0], x[:, 1], c=y)

ipywidgets.interact(plot_moons, samples=[200, 500, 1000], noise=(0,2,0.02))
```

```python
# Animating
def plot_sin(start=0, end=30, factor=1, grid=False, plot_cos=False):
  x = np.linspace(start, end, (end-start)*10)
  y = np.sin(x)*factor
  plt.grid(grid)
  plt.plot(x, y)
  if plot_cos:
    y = np.cos(x)
    plt.plot(x, y)

ipywidgets.interact(plot_sin, start=(0,10,1), end=(20,50,1), factor=(0,5,0.1), grid=False)
```

```python
# Sigmoid
import math
import ipywidgets

def plot_sigmoid(x_in=0):
  x = np.linspace(-5, 5, 1000)
  y = 1 / (1 + np.exp(-x))
  y_in = 1 / (1 + math.exp(-x_in))

  plt.plot(x,y)
  plt.scatter(x_in, y_in, c="r")

ipywidgets.interact(plot_sigmoid, x_in=(-5, 5, 0.1))
```

```python
# Histogram
import math
import ipywidgets

def plot_hist(mu=0, sigma=1, n=100, bins=10, color="pink"):
  plt.xlim(-20,20)
  x = np.random.normal(mu, sigma, n)
  plt.hist(x, bins=bins, colot=color)

ipywidgets.interact(plot_hist, mu=(-10,10,0.5), sigma=(0,10,0.1), n=(10,1000,1), bins=(1,100,1), color=["purple", "green", "blue"])
```
